```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

def extract_article_text(url,df1):
    try:
        # Send a HEAD request to check the status code
        response = requests.head(url)
        if response.status_code == 404:
            print(f"Error 404: {url} not found. Skipping.")
            return None, None

        # Send a GET request to the URL
        response = requests.get(url)
        response.raise_for_status()  # Check for errors in the response

        # Parse HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract article title and text
        title = soup.title.text.strip() if soup.title else 'No Title'
        text = ' '.join([p.text.strip() for p in soup.find_all('p')])

        return title, text
    except Exception as e:
        print(f"Error processing {url}: {e}")
        return None, None

def main():
    # Read the Excel file
    df = pd.read_excel('input.xlsx')
    df1=pd.DataFrame(df)
   

    # Iterate through rows and process each URL
    for index, row in df.iterrows():
        url = row['URL']
        title, text = extract_article_text(url,df1)

        if title and text:
            # Get the value of 'URL_ID' column to name the output file
            url_id = str(row['URL_ID'])
            
            # Create a text file based on 'URL_ID'
            output_filename = f'{url_id}.txt'

            # Write the title and text to the text file
            with open(output_filename, 'w', encoding='utf-8') as output_file:
                output_file.write(f"Title: {title}\n\n")
                output_file.write(f"Text: {text}\n\n")
                output_file.write("=" * 50 + "\n")
    df1.to_csv("RefinedInputs.csv")

if __name__ == "__main__":
    main()
```
